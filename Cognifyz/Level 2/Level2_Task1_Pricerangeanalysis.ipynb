**PRICE RANGE ANALYSIS**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Task/Dataset .csv')
```

```python
df[['Price range', 'Aggregate rating', 'Rating color']].head()
```

```python
df['Price range'].value_counts()
```

```python
avg_rating_price = df.groupby('Price range')['Aggregate rating'].mean()
avg_rating_price
```

```python
avg_rating_price.plot(kind='bar', figsize=(6,4))
plt.title("Average Rating by Price Range")
plt.ylabel("Average Rating")
plt.xlabel("Price Range")
plt.show()
```

```python
avg_rating_price.idxmax(), avg_rating_price.max()
```

```python
color_rating = df.groupby('Rating color')['Aggregate rating'].mean()
color_rating
```

```python
color_rating.idxmax(), color_rating.max()
```

Conclusion:
Price range analysis shows that certain price categories have higher average ratings.
The most common price range contains the highest number of restaurants.
Rating color analysis indicates which rating category represents better customer feedback.
