**FEATURE ENGINEERING**

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Task/Dataset .csv')
df.head()
```

```python
df[['Restaurant Name', 'Address', 'Has Table booking', 'Has Online delivery']].head()
```

```python
df['Restaurant_Name_Length'] = df['Restaurant Name'].str.len()
df[['Restaurant Name', 'Restaurant_Name_Length']].head()
```

```python
df['Address_Length'] = df['Address'].str.len()
df[['Address', 'Address_Length']].head()
```

```python
df['Has_Table_Booking'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})
df[['Has Table booking', 'Has_Table_Booking']].head()
```

```python
df['Has_Online_Delivery'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})
df[['Has Online delivery', 'Has_Online_Delivery']].head()
```

```python
df[['Restaurant_Name_Length', 'Address_Length',
    'Has_Table_Booking', 'Has_Online_Delivery']].describe()
```

Conclusion:
New features were created from existing columns to enhance the dataset.
The length of restaurant names and addresses were extracted as numerical features.
Categorical variables related to table booking and online delivery were encoded
into binary values to make them suitable for further analysis and modeling.
