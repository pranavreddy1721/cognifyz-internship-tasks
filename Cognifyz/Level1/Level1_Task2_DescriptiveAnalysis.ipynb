Dataset Overview:
The dataset contains information about restaurants, including location, cuisines, and ratings.
It includes both numerical and categorical variables used for analysis.


```python
df.describe()
```

Numerical Analysis Summary:
Basic statistical measures such as mean, median, and standard deviation were calculated
to understand the distribution of numerical features in the dataset.


```python
df.select_dtypes(include='number').agg(['mean','median','std'])
```

```python
df['Country Code'].value_counts()
```

```python
df['City'].value_counts().head(10)
```

```python
df['City'].value_counts().head(10).plot(kind='bar', figsize=(8,4))
plt.title("Top 10 Cities by Number of Restaurants")
plt.show()
```

```python
cuisines = df['Cuisines'].dropna().str.split(', ')
all_cuisines = cuisines.explode()

all_cuisines.value_counts().head(10)
```

```python
all_cuisines.value_counts().head(10).plot(kind='bar', figsize=(8,4))
plt.title("Top 10 Cuisines")
plt.show()
```

Categorical Analysis Summary:
The distribution of restaurants was analyzed based on country code, city, and cuisines.
The results highlight the cities with the highest number of restaurants and the most
commonly offered cuisines.


```python
print("Top City:", df['City'].value_counts().idxmax())
print("Top Cuisine:", all_cuisines.value_counts().idxmax())
```

Conclusion:
Descriptive analysis was performed to explore the structure and characteristics of the dataset.
